## Spoken Digit Classification

Out data consists of recordings of spoken digits in files at 8kHz. The recordings are trimmed so that they have near minimal silence at the beginnings and ends.

Import all libraries necessary in order to implement the classification we do:
* numpy
* matplotlib
* librosa
* pandas
* sklearn (for models, metrics, classification and confusion matrix)
* keras 

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import os 
import librosa.display
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, BatchNormalization
from keras.optimizers import SGD
from tensorflow.keras import optimizers
import pandas as pd
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import plot_confusion_matrix
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix
```

Upload the entire dataset,in a data structure.
With the object os, we specify the path where the dataset is placed.

```python
file = os.listdir('free-spoken-digit-dataset-master/recordings')
data=[]
for i in file:
    x , sr = librosa.load('free-spoken-digit-dataset-master/recordings/'+i)
    data.append(x)
```

In order to see what the system has uploaded, we can print just one element of
the data structure

```python
print(data[10])
```

Then extract STFT features from the audio. STFT (Short Term Fourier Transform) is a frequency feature representation for audio:

```python
%matplotlib inline
X = librosa.stft(data[10])
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(10, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz') 
plt.colorbar()
```

Input Signal to STFT features :

```python
X=[]
for i in range(len(data)):
    X.append(abs(librosa.stft(data[i]).mean(axis = 1).T))
X= np.array(X)
```

```python
print(X)
```

```python
y = [i[0] for i in file]
import pandas as pd 
Y = pd.get_dummies(y)
```

## Splitting Dataset
Split the dataset (75% training and 25% testing) into training and testing sets with STFT audio features as input, audio class as target label.

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25)
```

# Neural Network :

```python
model = Sequential()
model.optimizer = optimizers.Adam(learning_rate=0.0001)
model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(128, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
# sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy',
#               optimizer='rmsprop',
              optimizer='rmsprop',
              metrics=['accuracy'])
history = model.fit(X_train, y_train, 
          epochs = 300,
          batch_size = 64, 
          validation_data=(X_test, y_test),
          shuffle=True)
model.summary()
score = model.evaluate(X_test, y_test, batch_size=128)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('Epoch')
plt.ylabel('loss')
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.xlabel('Epoch')
plt.ylabel('accuracy')
```

```python
predict_label = np.argmax(model.predict(X_test),axis = 1)
d = np.array(y_test)
index_d = np.argwhere(d==1)
# print(type(y_test),y_test.shape,y_test)
# print(type(d),d.shape,d)
# print(type(predict_label),predict_label.shape,predict_label)
# print(type(index_d),index_d.shape,index_d)
# print(type(index_d[:,1]),index_d[:,1].shape,index_d[:,1])

cm = confusion_matrix(index_d[:,1], predict_label.tolist())

# cm = confusion_matrix(list(map(int,y_test)), predict_label.tolist())

plt.figure(dpi=100)
plt.imshow(cm,cmap=plt.cm.Blues)
tick_locs = np.arange(10)
ticks = ['{}'.format(i) for i in range(0, 10)]
plt.xticks(tick_locs, ticks)
plt.yticks(tick_locs, ticks)
plt.ylabel("True label")
plt.xlabel("Predicted label")
plt.title("Confusion matrix")


for first_index in range(len(cm)):    #rows
    for second_index in range(len(cm[first_index])):    #columns
        c = cm[second_index][first_index]
        if c!= 0:
            plt.text(first_index,second_index, c,color='red', fontsize=10, va='center', ha='center')
            
plt.show()
```
